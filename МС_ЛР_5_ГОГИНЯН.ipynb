**I.** Проверить гипотезу о равенстве математических ожиданий наблюдаемых
случайных величин при уровне значимости 0,06 разными способами ‒
непосредственным расчетом по формулам из Указания и с помощью функций
python.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as ss
from copy import deepcopy
    
def revread(f):
    return np.array(list(map(float, f.read().replace(',', '.').split())))

def show_distrib(distrib):       #
    for i in range(len(distrib)):
        print(distrib[i], end=",")
        if (i+1)%3==0:
            print(";")
    print()

def selav(sel): # x - selection
    sum=0
    for i in range(len(sel)):
        sum += sel[i]
    return sum/len(sel)

def squar_sel(sel):
    sum=0
    for i in range(len(sel)):
        sum += sel[i]**2
    return sum/len(sel)

def sel_unb_disp(sel): # Selective unbiased dispersion - Выборочная несмещенная дисперсия - S^2
    N = len(sel)
    av = selav(sel)
    disp = squar_sel(sel)
    return (N*(disp - av)/(N-1))

def Tcrit(selx, sely):
    N = len(selx)
    M = len(sely)
    return (selav(selx) - selav(sely)) * math.sqrt(M*N*(N+M-2)/(N+M)) / math.sqrt(sel_unb_disp(selx)*(N-1) + sel_unb_disp(sely)*(M-1))


f = open('DLW-5.txt') # выборка
distrib = revread(f)
show_distrib(distrib)

selection = [[], [], []]
for i in range(len(distrib)):
    if i%3 == 0:
        selection[0].append(distrib[i])
    elif i%3 == 1:
        selection[1].append(distrib[i])
    elif i%3 == 2:
        selection[2].append(distrib[i])

print("X=\n",selection[0])
print("\nAverage x: ", selav(selection[0]))
print("Disp x^2: ", squar_sel(selection[0]))
print("Selective unbiased dispersion S^2(x): ", sel_unb_disp(selection[0]))

print("\nY=\n",selection[1])
print("\nAverage y: ", selav(selection[1]))
print("Disp y^2: ", squar_sel(selection[1]))
print("Selective unbiased dispersion S^2(y): ", sel_unb_disp(selection[1]))

print("\nZ=\n",selection[2])
print("\nAverage z: ", selav(selection[2]))
print("Disp z^2: ", squar_sel(selection[2]))
print("Selective unbiased dispersion S^2(z): ", sel_unb_disp(selection[2]))

        # first and second
print("\nКритерий T_M,N(x, y) = ", Tcrit(selection[0], selection[1]))

        # first and third
print("Критерий T_M,N(x, z) = ", Tcrit(selection[0], selection[2]))

        # second and third
print("Критерий T_M,N(y, z) = ", Tcrit(selection[1], selection[2]))
```

```python
# 2 пункт ==========================================
def total_average(selection):   # Расчет общего среднего значения
    sum=0
    for i in range(len(selection)):
        for j in range(len(selection[i])):
            sum += selection[i][j]
    return sum / (len(selection)*len(selection[0]))

def Scommon(selection): # Расчет общей суммы квадратов отклонений
    utotal = total_average(selection)
    sum=0
    for i in range(len(selection)):
        for j in range(len(selection[i])):
            sum += (selection[i][j] - utotal)**2    
    return sum        
            
def Sfactor_sum(selection, u):  #  Расчет факторной суммы квадратов отклонений
    utotal = total_average(selection)
    sum=0
    for i in range(len(u)):
        sum += (u[i] - utotal)**2
    return sum*len(selection[0])
        
def Sresidual_sum(selection, u):   # Расчет остаточной суммы квадратов отклонений
    sum=0
    for i in range(len(selection)):
        for j in range(len(selection[i])):
            sum += (selection[i][j] - u[i])**2    
    return sum            

    

# Расчет общего среднего значения
# utotal = total_average(selection)

# Расчет групповых средних
u = []
for i in range(len(selection)):
    u.append(selav(selection[i]))
    
# Расчет общей суммы квадратов отклонений
print("Расчет общей суммы квадратов отклонений: ", Scommon(selection)) 

# Расчет факторной суммы квадратов отклонений
print("Расчет факторной суммы квадратов отклонений: ", Sfactor_sum(selection, u))

# Расчет остаточной суммы квадратов отклонений
if round(Sresidual_sum(selection, u), 5) == round(Scommon(selection) - Sfactor_sum(selection, u), 5):
    print("Расчет остаточной суммы квадратов отклонений:", Sresidual_sum(selection, u))
else:
    print("\nГде-то косяк\n")
    print("Расчет остаточной суммы квадратов отклонений напрямую:", Sresidual_sum(selection, u))
    print("Scommon - Sfactor_sum =", Scommon(selection) - Sfactor_sum(selection, u))
    
# Расчет значения критерия F_N,m
k1 = len(selection)-1
k2 = len(selection)*(len(selection[0])-1)
Sfactor_sum2 = Sfactor_sum(selection, u)/(k1)
print("k1 =", k1, "\nk2 =", k2)

print("\nРасчет факторной суммы квадратов отклонений в квадрате: ", Sfactor_sum2)
Sresidual_sum2 = Sresidual_sum(selection, u)/(k2)

print("\nРасчет остаточной суммы квадратов отклонений в квадрате: ", Sresidual_sum2)

Fcrit = Sfactor_sum2 / Sresidual_sum2
print("\nРасчет значения критерия F_N,m :", Fcrit)
```

**II.** Проверить гипотезу о равенстве дисперсий наблюдаемых случайных
величин при уровне значимости 0,06 разными способами ‒ непосредственным
расчетом по формулам из Указания и с помощью функций python.

```python
S12 = lambda Sx1, Sx2: max(Sx1, Sx2)

S22 = lambda Sx1, Sx2: min(Sx1, Sx2)

def F(Sx1, Sx2):
    return S12(Sx1, Sx2)/ S22(Sx1, Sx2)

print("\nРасчет значения критерия F_N,m ")

for i in range(len(selection)):
    for j in range(len(selection)):
        if i<j:
            print("("+str(i+1)+","+str(j+1)+")", end=": ")
            print("S12("+str(i+1)+","+str(j+1)+")=", S12(sel_unb_disp(selection[i]), sel_unb_disp(selection[j])), end="; ")
            print("S22("+str(i+1)+","+str(j+1)+")=", S22(sel_unb_disp(selection[i]), sel_unb_disp(selection[j])), end="; ")
            print("k1 =", len(selection[0])-1, " k2 =", len(selection[0])-1, end="; ")
            print("F_N,M("+str(i+1)+","+str(j+1)+")=", F(sel_unb_disp(selection[i]), sel_unb_disp(selection[j])))
```
