```python
# Вычислить определенный интеграл функции y = x*sqrt(49 - x^2) на отрезке [-2, 2] методами прямоугольников, 
# трапеций и Симпсона с постоянным шагом h = 1.0
```

```python
import math

def f(x): # функция
    if x>=-7 and x<=7:
        return x*math.sqrt(49 - x**2)
    else:
        return 0

def trueint(a, b):   # значение интеграла в лоб
    Fa = (-1/3) * (49 - a**2)**(3/2)
    Fb = (-1/3) * (49 - b**2)**(3/2)
    return  Fb - Fa
    

# Метод левых прямоугольников
def leftrectangle(a, b, n = 4):
    sum = 0
    h = (b-a)/n
    print("Шаг:", h)
    x = a
    for i in range(n):
        # print("x= ", x, "x + h =", x+h, "f(x) = ", f(x))
        sum += f(x) * h   # сначала суммируем по левому иксу, потом прибавляем
        x = x + h
    return round(sum, 10)
        
# Метод правых прямоугольников
def rightrectangle(a, b, n = 4):
    sum = 0
    h = (b-a)/n
    print("Шаг:", h)
    x = a
    for i in range(n):
        # print("x= ", x, "x + h =", x+h, "f(x) = ", f(x))
        x = x + h        # сначала меняем х и получаем правую разность
        sum += f(x) * h
    return round(sum, 10)

# Формула средних прямоугольников
def midlrectangle(a, b, n = 4):
    sum = 0
    h = (b-a)/n
    print("Шаг:", h)
    x = a
    for i in range(n):
        # print("x= ", x, "x + h =", x+h, "f(x) = ", f(x))
        sum += f((x + (x + h)) / 2) * h
        x = x + h 
    return round(sum, 10)
    
# Метод трапеций
def trapezium(a, b, n = 4):
    sum = 0
    h = (b - a) / n
    print("Шаг:", h)
    x = a
    for i in range(n):
        # print("i = ", i, "x= ", x, "x + h =", x+h, "f(x) = ", f(x), "f(x+h) = ", f(x+h))
        sum += (f(x) + f(x + h)) * h / 2
        x = x + h
    return sum

def simpson(a, b, n = 2):
    n = n * 2     # тк н обязано быть четным
    h = (b - a) / n 
    print("Шаг:", h)
    x = a + h  # на начало ставим х1
    sum = (f(a) + f(b))
    for i in range(1, n):
        if i%2 == 0:  # если индекс Четный
            sum += 2 * f(x)
            x = x + h
        else:
            sum += 4 * f(x)
            x = x + h 
            
    sum = sum * (b - a) / (3 * n)
    return sum

a = -2.0
b = 2.0
n = 20
print("Значение интеграла = ", trueint(a, b), end="\n\n")
print("Метод левых прямоугольников = ", leftrectangle(a, b), end="\n\n")
print("Метод правых прямоугольников = ", rightrectangle(a, b), end="\n\n")
print("Метод средних прямоугольников = ", midlrectangle(a, b), end="\n\n")
print("Метод трапеций = ", trapezium(a, b), end="\n\n")
print("Метод Симпсона = ", simpson(a, b, 1), end="\n\n")
```
