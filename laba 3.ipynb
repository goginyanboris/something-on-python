```python
import numpy
import math

a = [[22, -3, -8, 7],[-8, -22, -4, -8],[8, -2, -18, 2],[7, 2, -9, -24]]
y = [-158, 254, -108, -24]


def show(a, y):
    for i in range(len(a)):
        s=""
        for j in range(len(a[i])):
            if a[i][j] >= 0 and j>0:
                s += " +"
            else:
                s += " "
            s += str(a[i][j]) + "x" + str(j)    
        s += " = " + str(y[i])
        print(s)
        
def inv_show(a, y):
    for i in range(len(a)):
        s= "x" + str(i) + " = " + str(y[i])
        for j in range(len(a[i])): 
            if a[i][j] >= 0:
                s += " + " + str(a[i][j])
            else:
                s += " " + str(a[i][j])
        print(s)

def showx(x):
    for i in range(len(x)):
        print("x"+str(i)+" =", x[i])
        

def inverse_m(a, y):
    inv_a = [[] for i in range(len(a))]
    inv_y = []
    for i in range(len(a)):
        for j in range(len(a[i])):
            if j != i :
                inv_a[i].append((-1)*a[i][j]/a[i][i])
        inv_y.append(y[i]/a[i][i])
    return inv_a, inv_y
        

def norm(a): # норма матрицы ||A||_1 - сумма по строкам
    normmas=[]
    for i in range(len(a)):
        normmas.append(0)
        for j in range(len(a[i])):
            #print(a[j][i])
            normmas[i]+=abs(a[i][j])
    return max(normmas)

def normy(x):
    summ=[]
    for i in range(len(x)):
        summ.append(abs(x[i]))
    #print("Норма x =", summ)
    return max(summ)


def simpl_it(a, y, n, e):
    x = [[0] for i in range(len(a))] #  массив последовательностей приближений к корням икс 
    bufx = [0 for i in range(len(a))]
    totalx = []
    # далее рекурентные формулы
    maximum = -1
    eps = (1-n)*e/n
    stop = eps + 100
    itcount=0
    while stop >= eps:
        itcount+=1
        k=0
        for i in range(len(x)): # цикл по иксам
            sum=0
            #print("x=", x, "\n======================")
            
            for j in range(len(a[i])): #цикл складывающий элементы
                buf=0
                if i==k%4:
                    k+=1
                for l in range(len(x)): # выбираем нужный х
                    if l==k%4:
                        buf = x[l][len(x[l])-1]     
                #print("k=", k)        
                k+=1        
                #print("a = ", a[i][j], " x[j]=", buf)
                sum += a[i][j]*buf
            
            bufx[i] = (y[i] + sum)
        for i in range(len(bufx)):
            x[i].append(bufx[i])
            bufx[i]=0
        # print(x)   - вывод приближений к корню         
        maxbuf = []
        for i in range(len(x)):
            maxbuf.append(abs(x[i][len(x[i])-1]-x[i][len(x[i])-2]))
        # print("MAXBUF\n", maxbuf) - вывод погрешности
        stop = max(maxbuf)
        #stop = maximum
    print("Кол-во итераций: ", itcount)   
    for i in range(len(x)):
        totalx.append(x[i][len(x[i]) - 1])
    normx=[]
    for i in range(len(x)):
        normx.append(x[i][2] - x[i][1])
    normx1 = normy(normx)
    print(normx1)
    return totalx

def seidel(a, y, n, e):
    x = [[0] for i in range(len(a))] #  массив последовательностей приближений к корням икс 
    totalx = []
    # далее рекурентные формулы
    maximum = -1
    eps = (1-n)*e/n
    stop = eps + 100
    itcount=0
    while stop >= eps:
        itcount+=1
        k=0
        for i in range(len(x)): # цикл по иксам
            sum=0
            #print("x=", x, "\n===================")
            for j in range(len(a[i])): #цикл складывающий элементы
                buf=0
                if i==k%4:
                    k+=1
                for l in range(len(x)): # выбираем нужный х
                    if l==k%4:
                        buf = x[l][len(x[l])-1]     
                #print("k=", k)        
                k+=1        
                #print("a = ", a[i][j], " x[j]=", buf)
                sum += a[i][j]*buf
            x[i].append((y[i] + sum))
        # print(x)   - вывод приближений к корню         
        maxbuf = []
        for i in range(len(x)):
            maxbuf.append(abs(x[i][len(x[i])-1]-x[i][len(x[i])-2]))
        # print("MAXBUF\n", maxbuf) - вывод погрешности
        stop = max(maxbuf)
    
    print("Кол-во итераций: ", itcount)   
    for i in range(len(x)):
        totalx.append(x[i][len(x[i]) - 1])
    return totalx

def evaluation(normm, normx, eps):
    print("\nНорма С = ", normm)
    print("Норма ||x_1 - x_0|| =", normx)
    print("C/(1-C) =", normm/(1-normm), " ===>")
    print("||x_k - x*|| <= ||x_1 - x_0|| * C^k/(1-C), где k=", 1)
    ans = (normm/(1-normm))*normx
    return ans

#========= main ===================================

print("Исходная матрица")
show(a, y)

print("Приведем к виду x = D + Cx")
inv_a, inv_y = inverse_m(a, y)
#print(inv_a)

inv_show(inv_a, inv_y)

normmas = norm(inv_a)
print("\nНорма матрицы =", normmas)

normx =normy(inv_y)
print("Норма ||x|| = ", normx)

print("\nМетод простых итераций")
x = simpl_it(inv_a, inv_y, normmas, 0.01)
showx(x)

print("\nМетод Зейделя")
z = seidel(inv_a, inv_y, normmas, 0.01)
showx(z)

normx1 = 5.306818181818182
print("normx1 =", 5.306818181818182)
k = evaluation(normmas, normx1, 0.01)
print("Погрешность первой итерации : |x_1 - x_0|| * C/(1-C) = ", k)



```
