```python
# построить интерполяционные полиномы Лагранжа и Ньютона в заданных
# точках у = x^2*e^x; x0=-1.2, x1= -0.7, x2= -0.2, x3= 0.3
# Вычислить значение в точке x = -0.5
# Построить графики полученных многочленов и исходной функции

import numpy as np 
import matplotlib.pyplot as plt
import math

def lagrange(x, y, t): 
    sum = 0 
    for j in range(len(y)): # счетчик суммы членов
        p1 = 1 
        p2 = 1 
        for i in range(len(x)): # счетчик умножкния
            if i != j:
                p1 = p1 * (t-x[i]) 
                p2 = p2 * (x[j]-x[i]) 
        sum = sum + y[j] * p1/p2 
    return sum

def coef(x, y): 
    #x : array of data points 
    #y : array of f(x) 
    n = len(x) 
    a = [] 
    for i in range(n): 
        a.append(y[i]) 

    for j in range(1, n): #
        for i in range(n-1, j-1, -1):  #start on n-1; go to j-1 with step -1
            a[i] = float(a[i]-a[i-1])/float(x[i]-x[i-j]) 

    return np.array(a) # return an array of coefficient 

def newton(x, y, xstar): 
    #a : array returned by function coef() 
    #x : array of data points 
    #y : array of f(x)
    #xstar : the node to interpolate at  
    a = coef(x, y)
    n = len(a) - 1 
    temp = a[n]
    print("coef = ", a)
    for i in range(n - 1, -1, -1): 
        temp = temp * (xstar - x[i]) + a[i] 
    return temp # return the y_value interpolation 


x=[-1.2, -0.7, -0.2, 0.3]
y = []
for i in range(4):
    y.append( (x[i]**2)*(math.exp(x[i])) )
xstar = -2.0

print('newton: ', newton(x, y, xstar))
print('lagrange: ', lagrange(x,y,xstar))
print('function: ', (xstar**2)*(math.exp(xstar)))
```

```python
f_x = np.arange(-4,5,0.01) 
f_y = [(i**2)*(math.exp(i)) for i in f_x] 
plt.plot(x, y,'o',f_x,f_y) 
plt.grid(True)
plt.ylim(-5, 10)
plt.xlim(-5,10)
plt.show()
plt.plot(x, y,'o',f_x,f_y) 
plt.grid(True)
plt.ylim(-0.5, 6)
plt.xlim(-3,2)
plt.show()
plt.plot(x, y,'o',f_x,f_y) 
plt.grid(True)
plt.ylim(-0.1, 0.5)
plt.xlim(-1.3,0.35)
plt.show()
```

```python
xn=np.linspace(x[0],x[3],100) 
yn=[lagrange(x,y,i) for i in xn] 
plt.title("Лагранж") # подпись графика
plt.xlabel("X")
plt.ylabel("Y") # подпись оси У
plt.plot(x,y,'o',xn,yn) 
plt.grid(True) 
plt.show()
plt.xlabel("X")
plt.ylabel("Y") # подпись оси У
plt.plot(x,y,'o',xn,yn) 
plt.grid(True) 
plt.ylim(-1, 10)
plt.xlim(-5,3)
plt.show()
```

```python
xn=np.linspace(x[0],x[3],100) 
yn=[newton(x,y,i) for i in xn] 
plt.title("Ньютон") # подпись графика
plt.xlabel("X")
plt.ylabel("Y") # подпись оси У
plt.plot(x,y,'o',xn,yn) 
plt.grid(True) 
plt.show()
plt.xlabel("X")
plt.ylabel("Y") # подпись оси У
plt.plot(x,y,'o',xn,yn) 
plt.grid(True)
plt.ylim(-1, 7)
plt.xlim(-3,3)
plt.show()
```
